```python
def get_url(city,month):
    url = "https://www.aqistudy.cn/historydata/daydata.php?city=" + city + "&month=" + month
    return url
```

```python
citys = ["北京","上海","广州","海口"]
years = ["2018","2019"]
months = ["0" + str(e) if len(str(e)) == 1 else str(e) for e in list(range(1,13))]
print(citys, years, months)
```

```python
urls = []
# for city in citys:
city = citys[3]
for year in years:
    for month in months:
        urls.append(get_url(city,year+month))
for url in urls:
    print(url)
```
